### TASK 6: Implementing a Decision Tree Classifier and Visualizing it graphically

##### Objective:
For the given ‘Iris’ dataset, building a decision tree classifier to predict the new data to the right class. 


### Importing the Data Set

```python
import pandas as pd 
data=pd.read_csv("Iris.csv")
data.head()
```

### Splitting the data into Dependent and Independent Columns

```python
x=data.drop(columns=["Species","Id"])
y=data["Species"]
y.describe()
```

```python
data.isnull().sum()
```

### Fitting the data to Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtree= DecisionTreeClassifier(random_state=0)  # by default criterion is "gini"
dtree.fit(x,y)
```

### Visualizing the Decision Tree Algorithm to understand it better

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree,export_text
```

```python
plt.figure(figsize=(18,9))
plot_tree(dtree,feature_names=x.columns,filled="True")
```
